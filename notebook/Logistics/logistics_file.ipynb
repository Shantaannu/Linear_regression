### Logistic Regression ‚Äî Summary

* Logistic regression is a binary classifier from the Generalized Linear Model (GLM) family.

* It models the probability of a categorical outcome (0 or 1), specifically estimating P(Y = 1) based on input variables 
ùëã
X.

* Unlike linear regression (used for continuous outcomes), logistic regression is used when the dependent variable is categorical.

### Key Assumptions
* The dependent variable must be binary (0/1).

* The value 1 should represent the desired outcome.

* Include only relevant, meaningful predictors.

* Low multicollinearity among independent variables.

* Independent variables must be linearly related to the log-odds of the outcome.

* Requires a relatively large sample size.
